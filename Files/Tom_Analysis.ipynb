```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_Stockdata = pd.read_csv("masterdata_Tom.csv", encoding= 'unicode_escape')
asx_merge = df_Stockdata[["DATE","FEDFUNDS","S&P/ASX 200"]].set_index("DATE")
nasdaq_merge = df_Stockdata[["DATE","FEDFUNDS","NASDAQ"]].set_index("DATE")
tsx_merge = df_Stockdata[["DATE","FEDFUNDS","TSX"]].set_index("DATE")
ftse100_merge = df_Stockdata[["DATE","FEDFUNDS","FTSE100"]].set_index("DATE")
```

```python
asx_merge
fig, ax = plt.subplots()
fig.subplots_adjust(right=1)
fig.set_figheight(20)
fig.set_figwidth(30)

twin1 = ax.twinx()



p1, = ax.plot(asx_merge["S&P/ASX 200"], "b-", label="S&P ASX200")
p2, = twin1.plot(asx_merge["FEDFUNDS"],"r-", label="Fed_funds")

ax.set_xlim(0, 120)
ax.set_ylim(0, 10000)
twin1.set_ylim(0, 4)


ax.set_xlabel("Date")
ax.set_ylabel("S&P ASX200")
twin1.set_ylabel("Fed Funds Rate")

ax.yaxis.label.set_color(p1.get_color())
twin1.yaxis.label.set_color(p2.get_color())


ax.tick_params(axis='y', colors=p1.get_color())
twin1.tick_params(axis='y', colors=p2.get_color())

ax.tick_params(axis='x',rotation=45)

ax.legend(handles=[p1, p2])
```
