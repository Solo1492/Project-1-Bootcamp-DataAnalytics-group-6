```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_feddata = pd.read_csv("FEDFUNDS.csv")
df_n225 = pd.read_csv("N225.csv")
df_imoex = pd.read_csv("IMOEX.csv")
df_Sensex = pd.read_csv("Sensex_data.csv")
```

```python
Sensex_merge = df_Sensex.merge(df_feddata,left_on ="Date", right_on = "Date", how="inner")
imoex_merge = df_imoex.merge(df_feddata,left_on ="Date", right_on = "Date", how="inner").dropna()
n225_merge= df_n225.merge(df_feddata,left_on ="Date", right_on = "Date", how="inner")
```

```python
n225_merge2 = n225_merge[["Date","Close","Fed_funds"]].set_index("Date")

fig, ax = plt.subplots()
fig.subplots_adjust(right=1)
fig.set_figheight(20)
fig.set_figwidth(30)

twin1 = ax.twinx()



p1, = ax.plot(n225_merge2["Close"], "b-", label="n225")
p2, = twin1.plot(n225_merge2["Fed_funds"],"r-", label="Fed_funds")

ax.set_xlim(0, 120)
ax.set_ylim(0, 35000)
twin1.set_ylim(0, 4)


ax.set_xlabel("Date")
ax.set_ylabel("N225")
twin1.set_ylabel("Fed Funds Rate")

ax.yaxis.label.set_color(p1.get_color())
twin1.yaxis.label.set_color(p2.get_color())


ax.tick_params(axis='y', colors=p1.get_color())
twin1.tick_params(axis='y', colors=p2.get_color())

ax.tick_params(axis='x',rotation=45)

ax.legend(handles=[p1, p2])
```
