```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
master_csv= "Files/masterdata_Tom.csv"
```

```python
master_data = pd.read_csv(master_csv, encoding="unicode_escape")
master_data.dropna()
print(master_data.head())
print(master_data.columns)
```

```python
master_data.columns = master_data.columns.str.strip()
print(master_data.columns)
```

```python
#Fed fund rate plot
master_data.plot('DATE','FEDFUNDS', title = 'FEDFUNDS',figsize=(15,6))
```

```python
#S&P/ASX 200 rate plot
master_data.plot('DATE','S&P/ASX 200', title = 'S&P/ASX 200',figsize=(15,6))
```

```python
#FTSE100 rate plot
master_data.plot('DATE','FTSE100', title = 'FTSE100',figsize=(15,6))
```

```python
#NASDAQ rate plot
master_data.plot('DATE','NASDAQ',figsize=(15,6), title = 'NASDAQ')
```

```python
#TSX rate plot
master_data.plot('DATE','TSX',figsize=(15,6), title = 'TSX')

```

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)

plt.plot(master_data['TSX'], label = 'TSX')
plt.plot(master_data['DATE'],master_data['NASDAQ'], label = 'NASDAQ')
plt.plot(master_data['DATE'],master_data['FTSE100'], label = 'FTSE100')
plt.plot(master_data['DATE'],master_data['S&P/ASX 200'], label = 'S&P/ASX 200')
plt.legend()
plt.gca().get_xaxis().set_visible(False)
plt.gca().get_yaxis().set_visible(False)

```
